```python
short=df.query('runtime <{}'.format(100))
medium=df.query('runtime <{}'.format(200))
long=df.query('runtime >{}'.format(200))
```

```python
mp_short=short ['popularity'].mean()
mp_medium=medium['popularity'].mean()
mp_long=long['popularity'].mean()
```

```python
location=[1,2,3]
values = [mp_short,mp_medium,mp_long]
```

```python
labels=['short','medium','long']
plt.bar(location,values,tick_label=labels)
plt.title('average popularity by different run time(length)')
plt.xlabel('runtime')
plt.ylabel('average popularity')
```

```python
def extract_column(column_name):
    data=df[column_name].str.cat(sep='|')
    data=pd.Series(data.split('|'))
    count=data.value_counts(ascending=False)
    return count
```

```python
geners_count=extract_column('genres')
```

```python
geners_count
```

```python
df_genre_count=pd.DataFrame({'genres':geners_count.index,'count':geners_count.values})
f,ax=plt.subplots(figsize=(20,10))
sns.barplot(x='count',y='genres',data=df_genre_count)
ax.set_title('occurance per generes')
ax.set_xlabel('occurance')
ax.set_ylabel('genres')
plt.show
```
